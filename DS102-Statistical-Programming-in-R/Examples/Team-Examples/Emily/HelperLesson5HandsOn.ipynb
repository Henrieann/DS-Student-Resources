## Examples for Lesson 5 HandsOn DSO102 Solution

Similar to Lesson 5 HandsOn. Plotting a histogram, box plot and checking the normal distrubution with color and titles.

```r
Rivers1 = data.frame(rivers)
```

```r
library("ggplot2")
```

```r
Rivers2 <- ggplot(Rivers1, aes(y = rivers))
Rivers2

Rivers2 + geom_histogram(binwidth=900, fill="skyblue", color="white")
 + ggtitle("Histogram of River Lengths in North America") + xlab("Length (in miles)") + ylab("Count")
```

## Now to get the boxplot, you need to change the aesthetic above from x to y

```r

Rivers2 + geom_boxplot(outlier.color="deepskyblue4", fill="Pink") + ylab("Length (in miles)") + xlab("")
```

## Normal distribution: 

```r
ggplot(Rivers1, aes(sample=rivers, color="orange")) + geom_qq()
```
