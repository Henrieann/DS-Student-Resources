#### Creating a Faker Instance

```python
fake = Faker("en_US")
Faker.seed(123)
students=[
    {
    "Name":fake.unique.name(),
    "DOB":fake.unique.date_of_birth(),
    "City":fake.unique.city()
    }
     for i in range(100)]
```

#### View your newly generated data as a Pandas Data Frame

```python
pd.DataFrame(students)
```

```python
from tabulate import tabulate
print(tabulate(students, headers="keys", tablefmt="fancy_grid", showindex=True, missingval="N/A"))
```
