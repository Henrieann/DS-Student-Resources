# Merging Datasets in Python

```python
performances = pd.read_excel('../Data/performances.xlsx')
judgesAspectsUnique = pd.read_excel('../Data/judgesAspectsUnique.xlsx')
```

```python
performances.head()
```

```python
judgesAspectsUnique.head()
```

## Merging

```python
IceSkating = pd.merge(judgesAspectsUnique, performances)
```

```python
IceSkating.head()
```

## specifying the common variable

```python
IceSkating = pd.merge(judgesAspectsUnique, performances, on='performance_id')
```
