# Data Visualization in Python-Diamonds

```python
# Import packages
import pandas as pd
import seaborn as sns
```

```python
# Load data
df = sns.load_dataset('diamonds')
```

```python
# Inspect data
df.head()
```

```python
# Identify numeric cols vs categories
df.info()
```

```python
# Stats for numeric columns
df.describe()
```

## Histograms & Barplots

```python
# Histogram of carat
df.carat.hist()
```

```python
# Barplot of cut
df.groupby('cut').size().plot(kind='bar')
```

```python
# Barplot of color
df.groupby('color').size().plot(kind='bar')
```

```python
# Barplot of clarity
df.groupby('clarity').size().plot(kind='bar', rot=30)
```

```python
# Check remaining columns
df.columns
```

```python
# Histogram of depth
df.depth.hist()
```

```python
# Histogram of table
df.table.hist()
```

```python
# Histogram of price
df.price.hist()
```

```python
# Histogram of x
df.x.hist()
```

```python
# Historam of y
df.y.hist()
```

```python
# Histogram of z
df.z.hist()
```

## Correlation Heatmap

```python
# Correlation heatmap
sns.heatmap(df.corr(), cmap='Blues')
```

## Boxplots

```python
# Boxplot all
df.boxplot()
```

```python
# List cols for reference
df.columns
```

```python
# Create subset of cols to plot
bp_cols = ['carat', 'depth', 'table', 'x', 'y', 'z']
```

```python
# Plot boxplot with subset
df.boxplot(column=bp_cols)
```

```python
# Boxplot price
df.boxplot(column=['price'])
```

## Scatterplots

```python
# Scatterplot x and y
sns.scatterplot(data=df, x='x', y='y')
```

```python
# Scatterplot of z and y
sns.scatterplot(data=df, x='z', y='y', hue='color')
```

```python
# Scatterplot of carat and price
sns.scatterplot(data=df, x='carat', y='price', hue='clarity')
```

```python
# Scatterplot of depth and table
sns.scatterplot(data=df, x='depth', y='table', hue='cut')
```

## Violin Plots

```python
# Violin plot of cut and price
sns.violinplot(data=df, x='cut', y='price')
```

```python
# Violin plot of clarity and price
sns.violinplot(data=df, x='clarity', y='price')
```

```python
# Violin plot of color and price
sns.violinplot(data=df, x='color', y='price')
```
