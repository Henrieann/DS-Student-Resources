# Overview

Project planning is the process that captures the big picture view of your project. Unlike Scrum and Kanban which are concerned with the small details of specific tasks that need to be completed, a project plan views the project in terms of whole features.


```c-lms
topic: Project Planning
```



A project plan is your main tool for answering big-picture questions about your project. This is often the tool that is used to discuss the different needs of the departments in a company.

- This is a long-term plan that helps you answer some of the big questions your application will face
- When should you start advertising?
- When should you hire a QA tester?
- Are there any bottlenecks that will cause your development to slow down?
- If you hired another developer, how would that change the release date?

## Gantt Charts

Being able to predict the future is very difficult, so it helps to break down the thing you are predicting into smaller parts. A Gantt chart is a way of seeing all your smaller pieces put together in a timeline so that you can see who is working on what and when.

![Gantt chart. It has several bars on it going horizontally. There is a date bar across the top. Each of the bars has a different length and some of them overlap with each other on the date bar.](media/gantt-chart.png)

Gantt charts are popular for their ability to see where tasks overlap. In the above screenshot, if different people are working on the top 3 tasks, that is ok because those are all done in parallel. If one person is supposed to be doing all of those your project probably has a problem.

Another aspect of Gantt charts that is helpful is that you can generally express the dependencies between tasks. In the above screenshot, tasks with the triple dots either follow or precede another task. By defining how tasks relate to each other, it's possible to determine the minimum possible time for part of your project to be complete. It's also easier in this style to spot a task that is depended on by multiple other tasks. Failure to make progress on that single task effectively blocks many other tasks. This is a crucial thing to identify in your project since you'll almost always want to divert resources to make sure a task with many dependencies finishes on time.

The sequence of tasks that have the potential to block others is the critical path for your project. There is no way to complete the project sooner than the time it takes to do those tasks plus any that follow since those following tasks are dependent on it. Identifying your critical path helps you target the high-risk tasks early and give you a much better chance that your project stays on schedule.

Gantt charts are not the only tool you should use for managing your project but is an important one at the overall project level. Gantt charts allow you to:

- See when tasks will be started relative to each other
- See who is allocated to which tasks so you can make sure no individual is needed on two things at the same time
- See if one task is ever blocked waiting for another to continue

# Liquid Planner

The tool you will use to illustrate the idea of a project plan is **[Liquid Planner](https://liquidplanner.com)**. It is not a tool for tracking bugs or small tasks, but helps you see the big picture of your project. It is especially good for answering "what if" questions since it can recalculate your estimated completion date automatically and adjust your timelines as you adjust priorities. Not all Gantt chart systems have this capability, and for some scenarios, it is not even something you would want to happen.


![Liquid Planner management screen. There are menu options for home, timesheets, my work, projects, people, dashboards, analytics, and help. The Projects tab is highlighted. The screen shows a series of projects listed, with the owner and the effort remaining to its right. On the far right of the screen, it shows a Gantt chart with bars going horizontally for each task and the dates across the top.](media/liquid1.png)

In this screenshot, you can see a project named `Project Management Course` that has several tasks. `Lesson 1` - `Lesson 5` are all dependent on the `Design Course Content` task. That means they cannot start until it has completed. This is indicated by the three blue dots.

The grey bar that appears on top of the timelines for `Lesson 1` - `Lesson 3` indicates an interruption from another task or event. In this case, it is the holiday of Labor Day. This pushes the completion date of those three lessons back 1 day. Liquid can account for holiday or vacation days and adjusts your estimated completion date automatically. This is extremely helpful for seeing the impact of a team member taking a vacation and if you will need to reassign other team members to make sure critical parts of your project can be completed.

---

## Estimates

The upper case E indicates when a task is estimated to be complete. This is an estimate because for each task you can put in an optimistic and pessimistic time estimate. This is an important part of project management, things rarely go according to plan, and building in some space can make your plan much more resilient to change.

![A horizontal bar from an Gantt chart that has a letter E in the middle of it.](media/liquid2.png)

The white bar indicates the earliest and latest completion dates based on the estimated time you entered. The more tasks a task depends on, the longer this bar will be. The `QA` task is dependent on all the `Lesson` tasks, which is dependent on the `Design Course Content` so it will naturally have the most variability in its completion date since changes in any of the previous courses will affect it.

---

## What-If Scenarios

In the previous example, there have been three people working on the course, and it has an estimated completion date of September 24th. What if you added another person to work on lesson 4?

![Liquid Planner management screen. There are menu options for home, timesheets, my work, projects, people, dashboards, analytics, and help. The Projects tab is highlighted. The screen shows a series of projects listed, with the owner and the effort remaining to its right. On the far right of the screen, it shows a Gantt chart with bars going horizontally for each task and the dates across the top. Many of the bars on the Gantt chart have the letter E on them.](media/liquid3.png)

As you can see, the estimated completion date has shifted to be earlier. It is now September 17th instead of September 24th. Lesson 4 is now in line with Lessons 1 - 3 as it is not blocked by DavidK working on Lesson 1 first. This capability makes it possible to weigh the value of having someone else work on this part of the project in order to be done sooner.

# Tracking time

Since Liquid planner will automatically adjust your tasks based on who is assigned and the dependencies you have set up, you must manually track the completion of the tasks you work on. Say DavidK works on the `Design course content` task for 10 hours. This would adjust the overall project to look like this.

![Screen from Liquid Planner management tool. The first column says project management course. The next column says has people's names. The next column has remaining effort and the last column has a date. To the right is a Gantt chart with horizontal bars on it and a calandar bar at the top. Many of the bars have the letter E on them.](media/liquid4.png)

As you can see the `Design course content` taskbar is much shorter now, this represents that the remaining time is smaller. The original estimate was 16-24 hours, 10 of which have been logged, so the new estimate is between 6 and 14 hours of work remains. If no work had been logged, tomorrow the project plan would show an estimated completion date of September 18th instead of September 17th since one day has passed and the same amount of work remains.

---

## Deadlines

Since the project time will naturally just push back when no time is logged, it is important to enter deadlines for different parts of your project. This will tell you when you are at risk of missing these dates and give you a chance to address it.

Now, return to the original scenario that is estimated to be completed on September 24th.

![Screen from Liquid Planner management tool. The first column says project management course. The next column says has people's names. The next column has remaining effort and the last column has a date. To the right is a Gantt chart with horizontal bars on it and a calandar bar at the top. Many of the bars have the letter E on them. ](media/liquid1.png)

If you enter a deadline of September 21st this project will now be at risk of being late.

![Screen from Liquid Planner management tool. The first column says project management course. The next column says has people's names. The next column has remaining effort and the last column has a date. To the right is a Gantt chart with horizontal bars on it and a calandar bar at the top. Many of the bars have the letter E on them. In the Gantt chart, the date bar has changed color, as have several of the horizontal bars.](media/liquid5.png)

The diamond shape on the timeline represents the deadline set for September 21st. As you can see it is at the very left edge of the white bar that represents the earliest to latest completion dates. This means the tasks must all complete in the optimistic time and not in the pessimistic time. It is still possible this project can complete on time, but there is a larger chance it will not. If you were to move the deadline sooner, to September 18th, the project would now be impossible to finish on time.

![Screen from Liquid Planner management tool. The first column says project management course. The next column says has people's names. The next column has remaining effort and the last column has a date. To the right is a Gantt chart with horizontal bars on it and a calandar bar at the top. Many of the bars have the letter E on them. In the Gantt chart, the date bar has changed color, as have several of the horizontal bars.](media/liquid6.png)

The deadline is before the first possible completion date. In either of these scenarios, if this part of your project is on the critical path, it is a good idea to allocate more resources to ensure that you can complete it on time. One way to do that would be to reassign a task to someone else.

If another person was assigned to `Lesson2` or `Lesson 3`, nothing would change, since those tasks are not blocking any others. `Lesson 4` and `Lesson 5` however are both assigned to the same person, so changing one of those to a different person would allow them to be worked on in parallel and thereby move up the completion date. Time to see what the timeline looks like if `Lesson 4` is reassigned.

![Screen from Liquid Planner management tool. The first column says project management course. The next column says has people's names. The next column has remaining effort and the last column has a date. To the right is a Gantt chart with horizontal bars on it and a calandar bar at the top. Many of the bars have the letter E on them. In the Gantt chart, the date bar has changed color, as has one of the horizontal bars.](media/liquid7.png)

The project is now at a very low risk. The estimated completion date is just before the deadline. The most pessimistic estimate would have the project completing on the same day as the deadline. In general you will generally want a bit more time to account for things going wrong, but in general this project has now moved from definitely over schedule to one that will very likely complete on time.
