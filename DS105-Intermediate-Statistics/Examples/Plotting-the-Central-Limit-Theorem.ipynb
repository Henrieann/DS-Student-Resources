# Plotting the Central Limit Theorem

Random samples from independent variables will tend towards a normal (Gaussian) distribution, even if the variables are non-normally distributed

[Central Limit Theorem](https://en.wikipedia.org/wiki/Central_limit_theorem)

Sampling a distribution by any means becomes "Normal" as the sample size grows. All we need is for the observations to be independent. We don't even care about the shape of the population distribution. This surprising fact was proved in a fairly general form in 1810 by Pierre-Simon Laplace, and caused quite a stir because it is so intuitive. It is called the Central Limit Theorem... Be careful though, not everything is normal.

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
```

```python
# create a variable with 2000 points
n = 2000

# now create two signals each with 2000 points
# this first signal will be a sine wave that is linearly spaced from 0 to 6pi and we want n of them
sig1 = np.sin(np.linspace(0,6*np.pi, n))

# numpy random numbers where we have a uniform distribution (n) of them, then we'll scale up * 2-1
# which will give us a uniform distribution that has an average of zero and a range from -1 to +1
sig2 = np.random.rand(n)*2-1

#lets combine them then plot them
combi = sig1 + sig2

# so let's plot the individual signals and their distributions
# to do that we need the subplot geometry to be 2 x 3 with a figsize of
# various, where the top row will correspond to the signals
# and the bottom row will correspond to the distribution using histograms
#fig, ax = plt.subplots(2,3,figsize=(10,4))
fig, ax = plt.subplots(2,3,figsize=(16,8))

# plot the sine wave
ax[0,0].plot(sig1)
ax[0,0].set_title('Sine wave')
# this is what the histogram of a sine wave looks like
ax[1,0].hist(sig1,100)

# plot the uniformly distributed set of numbers
ax[0,1].plot(sig2, '.', alpha=.3)
ax[0,1].set_title('Uniform numbers')
# this is what the histogram of a uniform distribution looks like
ax[1,1].hist(sig2,100)

# plot the combination which leans towards a gaussian distribution
ax[0,2].plot(combi)
ax[0,2].set_title('Combined signal')
ax[1,2].hist(combi,100)

# how do we get rid of the x ticks for all graphs
# apply vectorization or flattening using np.matrix.flatten(ax)
for i in np.matrix.flatten(ax):
    i.set_xticks([])
    i.set_yticks([])

plt.show()
```
